```python
#데이터 생성

import pandas as pd

food = pd.DataFrame({
    "메뉴":["김밥","비빔밥","자장면","우동","김밥","자장면","김밥"],        
    "가격":[3000,7000,5500,5500,3500,6000,4000],
    "분류":["한식","한식","중식","일식","한식","중식","한식"]
})
food
```

```python
food.info()
food.head() # 기본 5개 출력
food.tail(3) 
# 통계 요약
food.describe()  #수치형
food.describe(include='number')  #수치형

food.describe(include=object) #문자형

# 고유한 값의 갯수 - value_counts()
food["메뉴"].value_counts()
food["분류"].value_counts()



```

```python
#할인가 칼럼 추가 - 할인율 10%
discount = 0.1
food["할인가"] = food["가격"] * (1 - discount)
food

# 자료형 변환 -astype(자료형)
food["할인가"] = food["할인가"].astype("int64")
food
```

```python
# food 데이타를 csv 파일로 변환
import pandas as pd

food.to_csv("datas/food.csv", index=False)

# food 데이터 읽기
df = pd.read_csv("datas/food.csv")
df

```

```python
# 정렬 - 오름차순, 내림차순 정렬 -> sort_values()abs

df = df.sort_values("가격", ascending=True)
#inplace=True - 즉시 적용
df.sort_values("가격", ascending=True, inplace=True)
df

#가격을 높은 순으로 정렬- 내림차순
df.sort_values("가격", ascending=False, inplace=True)
df

#정렬 초기화 - sort_index()
df = df.sort_index()
df
```

```python
# 조건 검색
# 메뉴가 "자장면"인 음식 정보
result1 = (df["메뉴"] == "자장면")
print(df[result1])

print("===============================")

#가격이 4000원 미만인 음식 정보
result2 = df["가격"] < 4000
print(df[result2])

print("===============================")
# 메뉴가 '자장면' 이면서, 가역이 6000원 이상인 음식 정보
result3 = (df["메뉴"] == "자장면") & (df["가격"] >= 6000)
print(df[result3])
print("4===============================")

# 메뉴가 '자장면' 이거나 가역이 6000원 이상인 음식 정보
result4 = (df["메뉴"] == "자장면") | (df["가격"] >= 6000)
print(df[result4])

# 분류가 '중식'이 아닌 음식을 검색하시요
print("5===============================")
result5 = (df["분류"] != "중식")
print(df[result5])

#정렬 초기화 - sort_index()
#df = df.sort_index()
#df


```
