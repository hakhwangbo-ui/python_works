### 판다스 자료구조 - 시리즈 관리
#### 데이터가 순차적으로 나열된 1차권 배열의 형태를 가진다
#### #의 개수가 많을 수록 글꼴이 작아진다. - 마크다운(Markdown)  문서: 확장자(.md)

```python
import pandas as pd # 별칭
data = ["키보드", "마우스", "USB 메모리"]
index = [1, 2, 3]
# 객체 생성
s1 = pd.Series(data, index)
print(s1)

# s1의 속성
print(s1.index)
print(s1.values)
print(s1.shape) #데이터의 형태, 즉 1차원 3개의 요소 - 튜플(3.)

#요소 선택
print(s1[1])
print(s1[2])
print(s1[3])
```

```python
# 두번째 시리즈
data = [20000, 32000, 10000]
index = [1, 2, 3]

s2 = pd.Series(data, index)
print(s2)
```

### 데이터 프레임 : 행과 열로 이루어진 2차원표 형태이다.
#### 데이타 프레임의 data는 파이션의 딕셔너리 자료구조로 만든다.

```python
import pandas as pd
# 데이터프레임 자료 생성(객체:df)
df = pd.DataFrame({
    "name" : s1, #첫변째 시리즈
    "price": s2  #두번째 시리즈
})
df
#데이터프레임이 속성
print(df.columns)  # zkffja
print(df.shape)  #자료형태 - (3,2)

# 칼럼 검색
print("name")
print(df["name"])
print("--------")
print("price")
print(df["price"])
```
